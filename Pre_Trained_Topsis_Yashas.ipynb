```python
import pandas as pd
import numpy as np
import sys
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.preprocessing import MinMaxScaler

def normalize_matrix(matrix):
    scaler = MinMaxScaler()
    normalized_matrix = scaler.fit_transform(matrix)
    return normalized_matrix

def apply_topsis(matrix, weights, impacts):
    normalized_matrix = normalize_matrix(matrix)
    m, n = normalized_matrix.shape
    weighted_matrix = np.multiply(normalized_matrix, weights)
    ideal_positive = np.max(weighted_matrix, axis=0)
    ideal_negative = np.min(weighted_matrix, axis=0)

    for i in range(n):
        if impacts[i] == '-':
            ideal_positive[i], ideal_negative[i] = ideal_negative[i], ideal_positive[i]

    positive_distances = np.linalg.norm(weighted_matrix - ideal_positive, axis=1)
    negative_distances = np.linalg.norm(weighted_matrix - ideal_negative, axis=1)

    topsis_scores = negative_distances / (negative_distances + positive_distances)

    return topsis_scores



def plot_topsis_bar_chart(models, topsis_scores):
    model_names = ['bert-base-uncased', 'roberta-base', 'distilbert-base-uncased', 'xinet-base-cased', 'albert-base-v2']
    plt.figure(figsize=(10, 6))
    sns.barplot(x=model_names, y=topsis_scores, palette='viridis')
    plt.title('TOPSIS Scores for Each Model')
    plt.xlabel('Models')
    plt.ylabel('TOPSIS Score')
    plt.xticks(rotation=45, ha='right')
    plt.tight_layout()
    plt.show()

def topsis_analysis(datafile, weights, impacts):
    if len(weights) != len(impacts):
        print("Number of elements in weights and impacts should be the same.")
        sys.exit(0)

    data = pd.read_csv(datafile, skiprows=0, index_col=0)  # Skip the first row as header and set the first column as index
    models = data.index.tolist()

    matrix = data.values
    topsis_scores = apply_topsis(matrix, weights, impacts)

    # Add TOPSIS scores to the dataframe
    data['TOPSIS_Score'] = topsis_scores

    # Rank the models based on TOPSIS scores
    data['Rank'] = data['TOPSIS_Score'].rank(ascending=False)

    # Save results to a new CSV file
    output_file = "topsis_results.csv"
    data.to_csv(output_file)
    print(f"Results saved to {output_file}")

    # Display TOPSIS analysis results
    print("\nTOPSIS Analysis Results:")
    print(data[['TOPSIS_Score', 'Rank']])

    # Plot bar chart for TOPSIS scores
    plot_topsis_bar_chart(models, topsis_scores)

if __name__ == "__main__":
    # Example usage
    datafile = "data.csv"  # Replace with your data file
    criteria_weights = [0.3, 0.4, 0.2, 0.5, 0.6]  # Adjust weights based on your criteria
    criteria_impacts = ['+', '-', '+', '-', '-']  # Adjust impacts based on your criteria

    topsis_analysis(datafile, criteria_weights, criteria_impacts)
```
